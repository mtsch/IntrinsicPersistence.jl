# Ko vse deluje

Najdlje traja samplanje 1M toƒçk.

```julia
using Plots; plotly()
using Manifolds
using IntrinsicPersistence
```

```julia
function lep_graf(man, r, n = 1_000_000)
    pts = rand(man, n)
    @time begin
        gc  = GeodesicComplex(pts, r)
        res = persistence(gc)
    end
    println(gc)
    plot(gc, res, graph = false)
end
```

```julia
# Torus
lep_graf(3Sphere(1) √ó Sphere(1), 0.3)
```

```julia
# Ukrivljena cev
lep_graf(Cube(1) √ó scaled(Sphere(1), x -> (x - 0.5)^2 + 1), 0.1)
```

```julia
# Navita zadeva
lep_graf(3reparametrized(Sphere(1), 2) √ó scaled(Sphere(1), exp), 0.5)
```

```julia
# Asimetriƒçen torus
lep_graf(4Sphere(1) √ó scaled(Sphere(1), x -> (x - 0.5)^2 + 1), 0.4)
```

```julia
# ƒåudno samplana cev
lep_graf(reparametrized(Cube(1), x -> x^2) √ó Sphere(1), 0.2)
```

```julia
# Malo toƒçk
lep_graf(3Sphere(1) √ó Sphere(1), 0.8)
```
